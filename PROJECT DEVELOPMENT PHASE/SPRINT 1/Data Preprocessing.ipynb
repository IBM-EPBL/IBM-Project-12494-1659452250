Importing the ImageDataGenerator class

```python
import keras
from keras.preprocessing.image import ImageDataGenerator
```

Defining parametes arguments for ImageDataGenerator class

```python
train_data = ImageDataGenerator(rescale= 1./255,shear_range= 0.2,
                                rotation_range= 180,horizontal_flip = True,
                                vertical_flip = True,zoom_range = 0.2)
```

```python
test_data = ImageDataGenerator(rescale= 1./255)
```

Applying ImageDataGenerator functionality to dataset

```python
#Applying ImageDataGenerator functionality to trainset
x_train = train_data.flow_from_directory(r"/content/drive/MyDrive/dataset/training_dataset",
                                         target_size = (128,128),
                                         class_mode = "binary",
                                         batch_size = 32)
```

```python
#Applying ImageDataGenerator functionality to testset
x_test = test_data.flow_from_directory(r"/content/drive/MyDrive/dataset/test_dataset",
                                       target_size = (128,128),
                                       class_mode = "binary",
                                       batch_size = 32)
```
